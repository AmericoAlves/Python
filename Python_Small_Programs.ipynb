```python
#Project - DNA Sample

sample = ['GTA','GGG','CAC']

def read_dna(dna_file):
    dna_data=""
    with open(dna_file,"r") as f:
        for line in f:
            dna_data+=line
    return dna_data

def dna_codons(dna):
    codons=[]
    for i in range (0,len(dna),3):
        if (i+3)<len(dna):
            codons.append(dna[i:i+3])
    return codons

def match_dna(dna):
    count=0
    for codon in dna:
        if codon in sample:
            count+=1
    return count

def is_criminal(dna_sample):
    dna_data=read_dna(dna_sample)
    codons=dna_codons(dna_data)
    num_count=match_dna(codons)
    if num_count>=3:
        print ("contagem: %d - Continuar a investigar!"%num_count)
    else:
        print ("Inocente")
        
        
is_criminal("S1.txt")
is_criminal("S2.txt")
is_criminal("S3.txt")
```

```python
'''
Project - Bank Account
'''
class BankAccount():
    balance=0
    def __init__(self,name):
        self.name=name
    def __repr__(self):
        return "Cliente: %s Saldo: %.2f"%(self.name,self.balance)
    def show_balance(self):
        print ("Saldo: %.2f"%self.balance)
    def deposit(self,amount):
        if amount<0:
            print ("valor impossvel")
        else:
            print ("Deposito: %.2f"%amount)
            self.balance+=amount
            self.show_balance()
    def withdraw(self,amount):
        if amount>self.balance:
            print ("Valor acima do limite!")
        else:
            print ("Levantamento: %.2f"%amount)
            self.balance-=amount
            self.show_balance()

minhaconta = BankAccount("Helder")
print (minhaconta)
minhaconta.show_balance()
minhaconta.deposit(-10)
minhaconta.deposit(1000)
minhaconta.withdraw(2500)
minhaconta.withdraw(500)
print (minhaconta)
    
```

```python
'''
Project - RGB & Hex Converter !
'''

def rgb_hex():
    invalid_msg="Not valid RGB!"
    print ("Usar valores entre 0 e 255!")
    red=int(input("Valor de vermelho: "))
    green=int(input("Valor de verde: "))
    blue=int(input("Valor de azul: "))
    if (red<0 or red>255)or(green<0 or green>255)or(blue<0 or blue>255):
        print (invalid_msg)
        return
    val = (red << 16) + (green << 8) + (blue)
    val = (hex(val)[2:]).upper()
    while len(val)<6:
        val = "0"+ val
    print ("%s" % val) 

def hex_rgb():
    hex_val=input("O hexadecimal: ")
    if len(hex_val) != 6:
        print ("Invalid Hex!")
        return
    else:
        hex_val=int(hex_val,16)
        two_hex_digits=2**8
        blue=hex_val%two_hex_digits
        hex_val = hex_val >> 8
        green=hex_val%two_hex_digits
        hex_val = hex_val >> 8
        red = hex_val%two_hex_digits
        print ("Red:%s Green:%s Blue:%s" % (red,green,blue))

def convert():
    while True:
        print ("Bem vindo ao conversor!")
        option=input("Enter: 1-RGB to Hex, 2-Hex to RGB, X to eXit: ")
        if option=="1":
            print ("RGB to hex...")
            rgb_hex()
        elif option =="2":
            print ("Hex to RGB...")
            hex_rgb()
        elif option=="X" or option=="x":
            break
        else:
            print ("Invalid!")

convert()
```

```python
'''
Project - Calendar
'''
from time import sleep, strftime
NAME = "Helder"
calendar={}

def welcome():
    print ("Bom dia %s"%NAME)
    print ("Vamos abrir o calendario")
    sleep (2)
    print ("Dia " + strftime("%A %B %d, %Y"))
    print ("Hora "+ strftime("%X"))
    sleep(2)
    print ("O que queres fazer? ")   
    
def start_calendar():
        welcome ()
        start = True
        while start:
            user_choice=input("Digite A para Add, U to Update, D para apagar, V to View e X to eXit:" )
            user_choice=user_choice.upper()
            if user_choice =="V":
                if len(calendar.keys())==0:
                    print ("O calendario esta vazio!")
                else:
                    print (calendar)
            elif user_choice=="U":
                date = input("Que data? ")
                update = input("Digite a atualizacao: ")
                calendar[date]=update
                print ("Atualizado!")
                print (calendar)
            elif user_choice=="A":
                event=input("Adicione nota: ")
                date=input("Adicione data (MM/DD/YYYY): ")
                if len(date)!=10 or int(date[6:])<int(strftime("%Y")):
                    print ("Data invalida")
                    resp=input("Outra data?Y/N: ")
                    resp = resp.upper()
                    if resp == "Y":
                        continue
                    else:
                        start=False
                else:
                    calendar[date]=event
                    print ("Adicionado")
                    print (calendar)
            elif user_choice=="D":
                if len(calendar.keys())==0:
                    print ("O calendario est vazio: ")
                else:
                    event=input("Qual nota?")
                    for date in calendar.keys():
                        if event==calendar[date]:
                            del calendar[date]
                            print ("Apagado")
                            print (calendar)
                        else:
                            print ("Essa nota nao existe")
            elif user_choice=="X":
                start=False
            else:
                print ("Invalido")
                start=False
                
start_calendar()                        
```

```python
"""
Project Rock, Papper, Scissors
Human vs Computer
"""
from random import randint

options = ["ROCK","PAPER","SCISSORS"]
message = {"tie":"Yawn it's a tie!","won":"Yay you won!","lost":"Aww you lost!"}

def decide_winner(user_choice,computer_choice):
    print ("A sua escolha foi %s!"%user_choice)
    print ("O computador escolheu %s"%computer_choice)
    if user_choice==computer_choice:
        print (message["tie"])
    elif user_choice==options[0] and computer_choice==options[2] or user_choice==options[1] and computer_choice==options[0] or user_choice==options[2] and computer_choice==options[1]:
        print (message["won"])
    else:
        print (message["lost"])

def play_RPS():
    user_choice=input("Enter Rock, Paper or Scissors: ")
    user_choice=user_choice.upper()
    computer_choice=options[randint(0,2)]
    decide_winner(user_choice,computer_choice)

play_RPS()
```

```python
"""
Roll the dices!
"""
from random import randint
from time import sleep

print ("Vamos testar a sua sorte!")

def get_user_guess():
    guess = int(input("Adivinhe o valor: "))
    return guess

def roll_dice(number_of_sides):
    first_roll=randint(1,number_of_sides)
    second_roll=randint(1,number_of_sides)
    max_val = 2*number_of_sides
    print ("For %d sides the maximum is %d!"%(number_of_sides,max_val))
    bet = get_user_guess()
    if bet>max_val:
        print ("Guess must be <= to %d!"%max_val)
    else:
        print ("Rolling...")
        sleep(2)
        print ("The first roll is %d!"%first_roll)
        sleep(1)
        print ("The second roll is %d!"%second_roll)
        sleep(1)
        total = first_roll+second_roll
        print ("Result..."),
        sleep(1)
        print (total)
        if bet==total:
            print ("You win!!!")
        else:
            print ("You lost, :-( ")

roll_dice (6)
print ("Game Over")
```

```python
# Battleship

from random import randint

#Board
board = []

for x in range(5):
    board.append(["O"] * 5)

def print_board(board):
    for row in board:
        print (" ".join(row))

print_board(board)

#Placement of the ship
def random_row(board):
    return randint(0, len(board) - 1)

def random_col(board):
    return randint(0, len(board[0]) - 1)

ship_row = random_row(board)
ship_col = random_col(board)

# Turns
for turn in range(4):
    print ("Turn", turn+1)
    guess_row = int(input("Guess Row: "))
    guess_col = int(input("Guess Col: "))

    if guess_row == ship_row and guess_col == ship_col:
        print ("Congratulations! You sunk my battleship!")
        break
    else:
        if (guess_row < 0 or guess_row > 4) or (guess_col < 0 or guess_col > 4):
            print ("Oops, that's not even in the ocean.")
        elif(board[guess_row][guess_col] == "X"):
            print ("You guessed that one already.")
        else:
            print ("You missed my battleship!")
            board[guess_row][guess_col] = "X"
    if turn==3:
        print ("Game Over")
    
    print_board(board)
```

```python
# Grades statistics

grades = [100, 100, 90, 40, 80, 100, 85, 70, 90, 65, 90, 85, 50.5]

def print_grades(grades_input):
    for grade in grades_input:
        print(grade)
print_grades(grades)

def grades_sum(scores):
    total = 0
    for score in scores: 
        total += score
    return total
print(grades_sum(grades))

def grades_average(grades_input):
    sum_of_grades = grades_sum(grades_input)
    average = sum_of_grades / float(len(grades_input))
    return average
print(grades_average(grades))

def grades_variance(scores):
    average=grades_average(scores)
    variance=0
    for score in scores:
        variance=variance+(average-score)**2
    return variance/len(scores)
print(grades_variance(grades))

def grades_std_deviation(variance):
    return variance**0.5
print(grades_std_deviation(grades_variance(grades)))
```

```python
# Students results

lloyd = {
  "name": "Lloyd",
  "homework": [90.0, 97.0, 75.0, 92.0],
  "quizzes": [88.0, 40.0, 94.0],
  "tests": [75.0, 90.0]
}
alice = {
  "name": "Alice",
  "homework": [100.0, 92.0, 98.0, 100.0],
  "quizzes": [82.0, 83.0, 91.0],
  "tests": [89.0, 97.0]
}
tyler = {
  "name": "Tyler",
  "homework": [0.0, 87.0, 75.0, 22.0],
  "quizzes": [0.0, 75.0, 78.0],
  "tests": [100.0, 100.0]
}

students = [alice,lloyd,tyler]

def average (numbers):
    total = float(sum(numbers))
    return total/len(numbers)

def get_average (student):
    homework = average (student["homework"])
    quizzes = average (student["quizzes"])
    tests = average (student["tests"])
    return homework*0.10+quizzes*0.30+tests*0.6

def get_letter_grade (score):
    if score>=90:
        return "A"
    elif 80<=score<90:
        return "B"
    elif 70<=score<80:
        return "C"
    elif 60<=score<70:
        return "D"
    else:
        return "F"

def get_class_average(class_list):
    results = []
    for student in class_list:
        results.append(get_average(student))
        return average (results)

print (get_letter_grade(get_average(tyler)))

print (get_class_average(students))

print (get_letter_grade (get_class_average(students)))
```

```python
# Shopping Cart and biling

shopping_list = ["banana", "orange", "apple"]

stock = {
  "banana": 6,
  "apple": 0,
  "orange": 32,
  "pear": 15
}
    
prices = {
  "banana": 4,
  "apple": 2,
  "orange": 1.5,
  "pear": 3
}

def compute_bill(food):
    total = 0
    for i in food:
        if stock[i]!=0:
            total = total + prices [i]
            stock[i]-= 1
    return total

print (compute_bill(shopping_list))
```

```python
# Classes

class Car(object):
    condition = "new"
  
    def __init__(self, model, color, mpg):
        self.model = model
        self.color = color
        self.mpg   = mpg
  
    def display_car (self):
        print ("This is a "+self.color+" "+self.model+" with " +str(self.mpg)+" MPG.")
  
    def drive_car(self):
        self.condition = "used"

my_car = Car("DeLorean", "silver", 88)

print (my_car.condition)
print(my_car.model)
print (my_car.color)
print (my_car.mpg)

my_car.drive_car()

print (my_car.condition)


# Inheritance
class ElectricCar(Car):
    
    def __init__(self,model, color, mpg, battery_type):
        super(ElectricCar, self).__init__(model, color, mpg) #super call
        self.battery_type = battery_type

# Override  
    def drive_car(self):
        self.condition="like new"

    
my_car = ElectricCar("Tesla","black",100,"molten salt")

print (my_car.condition)
print(my_car.model)
print (my_car.color)
print (my_car.mpg)
print (my_car.battery_type)

my_car.drive_car()

print (my_car.condition)
```

```python
# Trip cost

def hotel_cost (nights):
    return 140*nights

def plane_ride_cost(city):
    if city =="Charlotte":
        return 183
    elif city =="Tampa":
        return 220
    elif city == "Pittsburgh":
        return 222
    elif city == "Los Angeles":
        return 475

def rental_car_cost (days):
    if days >=7:
        return days*40-50
    elif days>=3 and days<7:
        return days*40-20
    else:
        return days*40

def trip_cost (city, days,spending_money):
    return hotel_cost(days-1)+plane_ride_cost(city)+rental_car_cost(days)+spending_money

print (trip_cost ("Los Angeles",5,600))
```

```python
# Scrabble

score = {"a": 1, "c": 3, "b": 3, "e": 1, "d": 2, "g": 2, 
         "f": 4, "i": 1, "h": 4, "k": 5, "j": 8, "m": 3, 
         "l": 1, "o": 1, "n": 1, "q": 10, "p": 3, "s": 1, 
         "r": 1, "u": 1, "t": 1, "w": 4, "v": 4, "y": 4, 
         "x": 8, "z": 10}

def scrabble_score (word):
    soma=0
    word = word.lower()
    for c in word:
        soma = soma + score[c]
    return soma
```
